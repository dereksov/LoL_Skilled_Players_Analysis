# Groupbys and Aggregates and Averages

```python
import pandas as pd
import numpy as np
import statsmodels.formula.api as smf   
## if error: python3.6 -m pip install scipy==1.2 --upgrade
import json 
import pprint
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import plotly.express as px
```

```python
players = pd.read_csv("data/NA/naplayer.csv")
```

```python
players["stats_kda"] = (players["stats_kills"] + players["stats_assists"]) / players['stats_deaths'].replace(0,1)
players["stats_gpm"] = players["stats_goldEarned"] / (players["gameDuration"]/ 60)
players["stats_dpm"] = players["stats_totalDamageDealtToChampions"] / (players["gameDuration"]/ 60)
```

```python
names = players["player_summonerName"].unique()
for name in names:
    print(name)
```

```python
c_interest = players.filter(like = "stats").columns
test_averages = players[players["player_summonerName"] == "Saiyans"][c_interest].mean()
```

```python
relevant = ["player_summonerName"] + list(c_interest)
```

```python
stats = players[relevant].groupby('player_summonerName').mean()
```

```python
stats["num_of_games"] = players[relevant].groupby('player_summonerName').count()["stats_champLevel"]
```

```python
stats["player_summonerName"] = stats.index
```
