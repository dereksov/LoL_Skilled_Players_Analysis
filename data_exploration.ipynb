```python
import pandas as pd
import numpy as np
import statsmodels.formula.api as smf   
## if error: python3.6 -m pip install scipy==1.2 --upgrade
import json 
```

```python
# Opening JSON file 

## http://ddragon.leagueoflegends.com/cdn/10.14.1/data/en_US/champion.json
## change the 10.14 to whatever the current patch is, for all champion data

with open('championids.json',  encoding='utf-8') as f:
    champion_data = json.load(f)

champion_data.keys()
champion_names = champion_data['data'].keys()
ids = []
for champion in champion_names:
    ids.append(champion_data['data'][champion]['key'])
ids = list(map(int, ids))

## Creates dictionary of champion id to champion name

champion_id_dict = dict(zip(ids, champion_names))
print(champion_id_dict)
```

```python
data = pd.read_csv("out.csv")
for col in data.columns:
    if len(data[col].unique()) ==1:
        data = data.drop(col,axis=1)
```

```python
columns = sorted(data.columns)
len(columns)
```

```python
# Exploring columns
for name in columns:
    print(name)
```

```python
# Encoding Losses to 0 

start_string = "participants_"
end_string = "_stats_win"

for i in range(0,10):
    key = start_string + str(i) + end_string
    data[key] = data[key].apply(lambda x: 0 if x == False else 1)
```

```python
data["teams_0_win"] = data["teams_0_win"].apply(lambda x: 0 if x == "Fail" else 1)
```

```python
champions = data[["gameId", "gameMode","teams_0_win","participants_0_championId", "participants_1_championId",
                    "participants_2_championId", "participants_3_championId", "participants_4_championId",
                    "participants_5_championId", "participants_6_championId", "participants_7_championId",
                    "participants_8_championId", "participants_9_championId","participants_0_stats_win",
                 "participants_1_stats_win", "participants_2_stats_win", "participants_3_stats_win",
                 "participants_4_stats_win","participants_5_stats_win","participants_6_stats_win",
                 "participants_7_stats_win","participants_8_stats_win","participants_9_stats_win",
                 "participants_0_timeline_lane", "participants_1_timeline_lane", "participants_2_timeline_lane",
                 "participants_3_timeline_lane","participants_4_timeline_lane","participants_5_timeline_lane",
                 "participants_6_timeline_lane","participants_7_timeline_lane","participants_8_timeline_lane",
                 "participants_9_timeline_lane",]].copy()
```

```python
# Remapping champion ids of the dataframe to champion names from dictionary

s_str = "participants_"
e_str = "_championId"
i = 0
for i in range(0,10):
    key = s_str + str(i) + e_str
    champions = champions.replace({key : champion_id_dict})
```

```python
# Boolean to remove ARAM games (you should use booleans,
# its fast than .drop() since drop uses booleans and indexing)
champions = champions[champions["gameMode"] != "ARAM"]
```
