# import, csv, json

```python
import pandas as pd
import numpy as np
import statsmodels.formula.api as smf   
## if error: python3.6 -m pip install scipy==1.2 --upgrade
import json 
import pprint
```

```python
df = pd.read_csv('data/clean.csv',low_memory=False)
print(df.shape)
```

```python
# Opening JSON file 

## http://ddragon.leagueoflegends.com/cdn/10.14.1/data/en_US/champion.json
## change the 10.14 to whatever the current patch is, for all champion data

with open('data/championids.json',  encoding='utf-8') as f:
    champion_data = json.load(f)
f.close()
```

```python
print(champion_data.keys())
```

```python
pprint.pprint(champion_data['data'])
```

```python
## Creates dictionary of champion id to champion name
champion_names = champion_data['data'].keys()
ids = [champion_data['data'][champion]['key'] for champion in champion_names]
ids = list(map(int, ids))
champion_id_dict = dict(zip(ids, champion_names))
pprint.pprint(champion_id_dict)
```

```python
# Remapping champion ids of the dataframe to champion names from dictionary

s_str = "participants_"
e_str = "_championId"
i = 0
for i in range(0,10):
    key = s_str + str(i) + e_str
    df = df.replace({key : champion_id_dict})
```

# regex + participants

```python
import re
```

```python
def strippart(num):
    res = []
    for name in df.columns:
        if re.search('^participants_{}_'.format(num), name) != None:
            res += [name]
    res = [re.sub('^participants_{}_'.format(num), '', x) for x in res]
    return res
```

```python
f0 = set(strippart(0))
f1 = set(strippart(1))
f2 = set(strippart(2))
f3 = set(strippart(3))
f4 = set(strippart(4))
f5 = set(strippart(5))
f6 = set(strippart(6))
f7 = set(strippart(7))
f8 = set(strippart(8))
f9 = set(strippart(9))
```

```python
len(f0),len(f1),len(f2),len(f3),len(f4),len(f5),len(f6),len(f7),len(f8),len(f9)
```

```python
set([x for x in f0 if 'playerScore' not in x])-f4
```

```python
common = list(f0 & f1 & f2 & f3 & f4 & f5 & f6 & f7 & f8 & f9)
```

```python
len(common)
```

```python
start = list(df.columns[:5])

team = []
for name in df.columns:
    if re.search('^team', name) != None:
        team += [name]

playercol = ['player_platformId', 'player_accountId', 'player_summonerName', 'player_summonerId', 
          'currentAccountId', 'player_matchHistoryUri', 'player_profileIcon']

partid = []
for name in df.columns:
    if re.search('^participantIdentities', name) != None:
        partid += [name]

def isolate(num):
    participant = ['participants_{}_{}'.format(num,x) for x in common]
    player = [x for x in partid if str(num) in x]
    res = df.copy()[start + team + participant + player]
    newcol = start + team + common + playercol
    res.columns = newcol
    res['participant_number'] = num
    return res
```

```python
i0 = isolate(0)
i1 = isolate(1)
i2 = isolate(2)
i3 = isolate(3)
i4 = isolate(4)
i5 = isolate(5)
i6 = isolate(6)
i7 = isolate(7)
i8 = isolate(8)
i9 = isolate(9)
```

```python
participants = pd.concat([i0,i1,i2,i3,i4,i5,i6,i7,i8,i9])
participants.shape
```
