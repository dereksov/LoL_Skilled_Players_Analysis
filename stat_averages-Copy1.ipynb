```python
def kda(row):
    if row["stats_deaths"] == 0:
        return (row["stats_kills"] + row["stats_assists"]) / 1
    else:
        return (row["stats_kills"] + row["stats_assists"]) / row["stats_deaths"]
```

```python
players["stats_kda"] = players.apply(lambda row: kda(row), axis = 1 )
players["stats_gpm"] = players["stats_goldEarned"] / (players["gameDuration"]/ 60)
players["stats_dpm"] = players["stats_totalDamageDealtToChampions"] / (players["gameDuration"]/ 60)
```
